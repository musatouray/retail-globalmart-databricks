### Create Medallion Architecture Catalogs
- bronze_dev
- silver_dev
- gold_dev

```python
%sql
--> create bronze catalog
CREATE CATALOG IF NOT EXISTS bronze_dev;

--> create silver catalog
CREATE CATALOG IF NOT EXISTS silver_dev;

--> create gold catalog
CREATE CATALOG IF NOT EXISTS gold_dev;
```

### Create Schemas

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze_dev.global_mart_retail
COMMENT 'Bronze Layer: Raw ingested data';

CREATE SCHEMA IF NOT EXISTS silver_dev.global_mart_retail
COMMENT 'Silver Layer: Cleansed and transformed data for downstream processing';

CREATE SCHEMA IF NOT EXISTS gold_dev.global_mart_retail
COMMENT 'Gold Layer: Enriched and aggregated data for business insights';
```

### Create Volume

```python
%sql

CREATE VOLUME IF NOT EXISTS bronze_dev.global_mart_retail.raw_data
COMMENT 'Volume for raw source files (CSV)';
```
